Display all prime numbers within an interval of 0 to 50.

```python
## Write your code here...

primes = [2]

for n in range(3, 51):
    isPrime = 0
    for i in range(2, ((n//2)+1)):
        if (n % i) == 0:
            isPrime = 1
    if isPrime == 0:    
        primes.append(n)
        isPrime = 1

print(primes)
```

### Matrix multiplication

NumPy has three alternative routines for multiplying two matrices, `m` and `n`.

- Element-wise multiplication: `numpy.multiply(m, n)` 
- Product of two matrices `numpy.matmul(m, n)` 
- Scalar/dot product `numpy.dot(m, n)` 
    
For matrix multiplication, you can also use `@` operator. For element-wise multiplication, you can also use `*` operator.

### Generation of periodic signals
In the following, we illustrate how to generate a periodic signal. 

```python
#Example
import numpy as np
import matplotlib.pyplot as plt

f = 1
a = np.linspace(0, 180, 180)
t=np.deg2rad(a)
signal = 0.4 * np.cos(2*np.pi*f*t) 
plt.plot(t, signal)
plt.show()
```

### Generating a complex signal

The code below demonstrates how to generate a complex signal. 

```python
#Example:
import numpy as np
import matplotlib.pyplot as plt
# Complex exponential signal
complex_exp_signal = np.exp(1j * t)  # 1j 

# Create a time vector
t = np.linspace(0, 2*np.pi, 500)

# Calculate complex exponential signal
complex_exp_signal = np.exp(1j * t)

# Draw a complex plane graph
plt.figure(figsize=(14, 6))

# Subfigure 1: Complex plane representation
plt.subplot(1, 2, 1)
plt.plot(np.real(complex_exp_signal), np.imag(complex_exp_signal)) 
plt.title('Complex Exponential Signal on Complex Plane')
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.grid(True)
plt.axis('equal')

# Sub-figure 2: Real and imaginary parts changing over time
plt.subplot(1, 2, 2)
plt.plot(t, np.real(complex_exp_signal), label='Real Part')
plt.plot(t, np.imag(complex_exp_signal), label='Imaginary Part', linestyle='--')
plt.title('Real and Imaginary Parts over Time')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# Display the graph
plt.tight_layout()
plt.show()
```

## Random variables and probability functions

The module of statistical funtions of SciPy contains a large number of probability distributions, summary and frequency statistics, correlation functions and statistical tests, masked statistics, kernel density estimation, quasi-Monte Carlo functionality, and more. Find more information at https://docs.scipy.org/doc/scipy/reference/stats.html

The codes below demonstrate how to generate samples from a uniform distribution and how to plot a probability density function (PDF) of a Normal distributed random variable.



```python
#Example:

import numpy as np
from scipy import stats
s = np.random.uniform(0,1,10) #take 10 samples from X ~ Uni(0,1)
print(s)
```

```python
#Example:
import numpy as np
#from numpy.fft import fft, fftfreq
from scipy.fftpack import fft, rfft, irfft, fftfreq, rfftfreq
import scipy.signal
from scipy.stats import norm, expon
import matplotlib.pyplot as plt

# Set the range of x
x = np.linspace(-4, 4, 1000)

# Probability density function of normal distribution (mean 0, standard deviation 1)
pdf_normal = norm.pdf(x, 0, 1)
mean_normal = norm.mean(0, 1)
var_normal = norm.var(0, 1)

# draw a normal distribution
plt.figure(figsize=(12, 6))
plt.plot(x, pdf_normal, label='PDF Normal')
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(True)
plt.legend()
plt.text(-3, 0.35, f'Mean: {mean_normal:.2f}\nVariance: {var_normal:.2f}')

plt.tight_layout()
plt.show()
```
